```python
X = df[['LogDistLeft', 'Categorical']]
y = df['LogTime']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.08, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
# print the intercept
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
prediction_test = lm.predict(X_train)
exppredictions = np.exp(predictions)
expy_test = np.exp(y_test)
```

```python
plt.scatter(expy_test,exppredictions,color='black', marker='*', alpha=0.7)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
#plt.scatter(y_train,prediction_test,color = 'red')
#plt.scatter(df['Distance Left'], df['Time to Busstop'])
```

```python
sns.distplot((expy_test-exppredictions),bins=10, color = 'black');
```

```python
plt.scatter(predictions, y_test-predictions)
#plt.scatter(prediction_test, y_train - prediction_test)
plt
```

```python
from sklearn import metrics

```

```python
print('MAE:', metrics.mean_absolute_error(expy_test, exppredictions))
print('MSE:', metrics.mean_squared_error(expy_test, exppredictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(expy_test, exppredictions)))
```
